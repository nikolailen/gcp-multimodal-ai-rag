```python
# Import necessary libraries
from google.cloud import bigquery

# Define a function to return the schema
schema = [
        bigquery.SchemaField("a_chunk_id", "STRING"),  # UUID format for chunk IDs
        bigquery.SchemaField("b_page_number", "INTEGER"),  # Page number
        bigquery.SchemaField("c_file_pages", "INTEGER"),  # Total number of pages in the file
        bigquery.SchemaField("d_document_date", "DATE"),  # Document date in YYYY-MM-DD format
        bigquery.SchemaField("e_chunk_type", "STRING"),  # Type of chunk, e.g., "text"
        bigquery.SchemaField("f_chunk_title", "STRING"),  # Title of the chunk
        bigquery.SchemaField("g_chunk_contents", "STRING"),  # Main content of the chunk
        bigquery.SchemaField("h_chunk_keywords", "STRING", mode="REPEATED"),  # List of keywords
        bigquery.SchemaField("i_chunk_summary", "STRING"),  # Summary of the chunk
        bigquery.SchemaField("j_file_number", "INTEGER"),  # File number
        bigquery.SchemaField("k_file_hash", "STRING"),  # Hash of the file
        bigquery.SchemaField("l_file_name", "STRING"),  # File name
        bigquery.SchemaField("m_file_layout", "STRING"),  # Layout of the file, e.g., "portrait"
        bigquery.SchemaField("n_document_name", "STRING"),  # Document name
        bigquery.SchemaField("o_document_author", "STRING"),  # Author of the document
        bigquery.SchemaField("p_document_type", "STRING"),  # Type of document, e.g., "Publication"
        bigquery.SchemaField("q_document_keywords", "STRING", mode="REPEATED"),  # List of document-level keywords
        bigquery.SchemaField("r_document_summary", "STRING"),  # Summary of the document
        bigquery.SchemaField("s_page_gsutil", "STRING"),  # Cloud storage URI for the page
        bigquery.SchemaField("t_page_url", "STRING"),  # URL for the page
        bigquery.SchemaField("u_file_gsutil", "STRING"),  # Cloud storage URI for the file
        bigquery.SchemaField("v_file_url", "STRING"),  # URL for the file
        bigquery.SchemaField("w_file_session_start", "DATETIME"),  # Start timestamp of the file session
        bigquery.SchemaField("x_file_session_end", "DATETIME"),  # End timestamp of the file session
        bigquery.SchemaField("y_file_session_duration", "TIME"),  # Duration of the file session
        bigquery.SchemaField("z_file_session_status", "STRING"),  # Status of the file session, e.g., "done"
        bigquery.SchemaField("aa_document_session_id", "STRING"),  # UUID for the document session
        bigquery.SchemaField("ab_document_session_start", "DATETIME"),  # Start timestamp of the document session
        bigquery.SchemaField("ac_document_session_end", "DATETIME"),  # End timestamp of the document session
        bigquery.SchemaField("ad_document_session_duration", "TIME"),  # Duration of the document session
        bigquery.SchemaField("ae_chunks_session_id", "STRING"),  # UUID for the chunks session
        bigquery.SchemaField("af_chunks_session_start", "DATETIME"),  # Start timestamp of the chunks session
        bigquery.SchemaField("ag_chunks_session_end", "DATETIME"),  # End timestamp of the chunks session
        bigquery.SchemaField("ah_chunks_session_duration", "TIME"),  # Duration of the chunks session
        bigquery.SchemaField("ai_document_finish_reason", "STRING"),  # Reason for document finish, e.g., "STOP"
        bigquery.SchemaField("aj_chunks_finish_reason", "STRING"),  # Reason for chunks finish, e.g., "STOP"
        bigquery.SchemaField("ak_document_prompt_token_count", "INTEGER"),  # Token count for document prompts
        bigquery.SchemaField("al_chunks_prompt_token_count", "INTEGER"),  # Token count for chunk prompts
        bigquery.SchemaField("am_document_candidates_token_count", "INTEGER"),  # Token count for document candidates
        bigquery.SchemaField("an_chunks_candidates_token_count", "INTEGER"),  # Token count for chunk candidates
        bigquery.SchemaField("ao_document_total_token_count", "INTEGER"),  # Total token count for the document
        bigquery.SchemaField("ap_chunks_total_token_count", "INTEGER"),  # Total token count for the chunks
        bigquery.SchemaField("aq_document_model_version", "STRING"),  # Version of the document processing model
        bigquery.SchemaField("ar_chunks_model_version", "STRING"),  # Version of the chunks processing model
    ]


# Initialize BigQuery client
client = bigquery.Client()

# Specify your dataset and table IDs
dataset_id = "your-project-id.KNOWLEDGE_BASE"  # Replace with your project.dataset
table_name = "CHUNKS"  # Replace with your table name
table_id = f"{dataset_id}.{table_name}"

# Create a Table object with the schema
table = bigquery.Table(table_id, schema=schema)

# Create the table in BigQuery
table = client.create_table(table, exists_ok=True)
print(f"Table {table_id} created or already exists.")
```
